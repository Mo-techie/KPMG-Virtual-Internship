```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
df = pd.read_excel('KPMG_VI_New_raw_data.xlsx', sheet_name='Transactions')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df_I = df[['transaction_id', 'product_id', 'customer_id', 'transaction_date', 'online_order', 'order_status', 'brand',
           'product_line', 'product_class', 'product_size', 'list_price', 'standard_cost', 'product_first_sold_date']]
```

```python
missing_rows_df = df_I[df_I.isnull().any(axis=1)]
missing_rows_df
```

```python
#investigating to see if the 197 missing values coincides
missing_columns = df[['brand', 'product_line', 'product_class', 'product_size', 'standard_cost', 'product_first_sold_date']]
missing_columns_df = missing_columns[missing_columns.isnull().any(axis=1)]
missing_columns_df
```

```python
#checking the missing values in online_order column
null_online_order = df.loc[df['online_order'].isnull()]
null_online_order
```

```python
#merging the two dataframes to check if missing values overlap
overlapped_missing_values = missing_columns_df.reset_index().merge(null_online_order.reset_index(), on=['index'])
overlapped_missing_values
```
